```python
# 1단계: 정점 갯수, 간선 갯수, 시작 정점 숫자 입력받기
N, M, V = map(int, input().split())

# 2단계: 리스트로 graph 정의
graph = [0]
for i in range(N):
  graph.append([]) #[0, [], [], [], []]

# 3단계: 정점과 간선 연결하기
for i in range(M):
  a, b = map(int, input().split())
  graph[a].append(b)
  graph[a].sort()
  graph[b].append(a)  # [0, [2,3],[1,5],[1,4],[3,5],[2,4]]
  graph[b].sort()

# 4단계: DFS 탐색 구현
def DFS(graph, v):
  if v in dfs:
    return
  else:
    dfs.append(v)
  for i in graph[v]:
    DFS(graph, i)

dfs = []
DFS(graph, V)
for i in dfs:
  print(i, end = ' ')
print()

# 5단계: BFS 탐색 구현
from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])
    visited.add(start)

    # 큐가 빌 때까지 반복
    while queue:
        # 큐에서 노드를 하나씩 꺼내어 출력
        node = queue.popleft()
        print(node, end=' ')

        # 현재 노드에 연결된 노드들 중 방문하지 않은 노드를 큐에 추가하고 방문 표시
        for neighbor in graph[node]:
            if neighbor not in visited:
                queue.append(neighbor)
                visited.add(neighbor)
bfs(graph, V)
```
